```python
maps_api_key = 'AIzaSyCtVNEK0BwvyHF2UdKTZ1oPdffC-ym8q5g'
endpoint = 'https://maps.googleapis.com/maps/api/directions/json?'

def get_distance_duration(origin, destination, mode = 'driving'):
    origin_str = f'{origin[0]},{origin[1]}'
    destination_str = f'{destination[0]},{destination[1]}'
    request_url = f'{endpoint}origin={origin_str}&destination={destination_str}&mode={mode}&key={maps_api_key}'
    response = requests.get(request_url)
    data = response.json()
    
    if data['status'] == 'OK':
        road_distance_meters = data['routes'][0]['legs'][0]['distance']['value']
        road_duration_seconds = data['routes'][0]['legs'][0]['duration']['value']

        return road_distance_meters, road_duration_seconds
        
    else:
        print('Error: Unable to calculate road distance and duration.')
        return None

get_distance_duration((-20.498489, -47.396929), (-20.487897, -47.416463))
```

```python
def grouped_api_call(cust_lat, cust_lng, cent_lat, cent_long):
    features = []
    for glat, glong, clat, clong in zip(cust_lat, cust_lng, cent_lat, cent_long):
        features.append(get_distance_duration((glat, glong),(clat, clong)))
    return features
```

```python
def haversine(latlon1, latlon2):
    lat1, lon1 = latlon1
    lat2, lon2 = latlon2
    R = 6371000  # radius of Earth in meters
    phi_1 = radians(lat1)
    phi_2 = radians(lat2)

    delta_phi = radians(lat2 - lat1)
    delta_lambda = radians(lon2 - lon1)

    a = (np.sin(delta_phi / 2) ** 2 +
         np.cos(phi_1) * np.cos(phi_2) * np.sin(delta_lambda / 2) ** 2)

    c = 2 * np.arctan2(np.sqrt(a), np.sqrt(1 - a))

    meters = R * c  # output distance in meters
    return meters/1000
```

```python
df[['CLUSTER_hdbscan', 'geolocation_lat','geolocation_lng', 'centroid_lat','centroid_lng']].groupby(by= ['CLUSTER_hdbscan']).apply(
    lambda row: grouped_api_call(row.geolocation_lat, row.geolocation_lng,row.centroid_lat,row.centroid_lng)
)
```

```python
for idx, row in df.iterrows():
    data = get_distance_duration((row.geolocation_lat, row.geolocation_lng),(row.centroid_lat,row.centroid_lng))
    if data:
        df.loc[idx,'road_distance'] = data[0]
        df.loc[idx,'road_duration'] = data[1]
```

```python
df['geo_distance'] = df[['CLUSTER_hdbscan', 'geolocation_lat','geolocation_lng', 'centroid_lat','centroid_lng']].apply(
    lambda row: haversine((row.geolocation_lat, row.geolocation_lng),(row.centroid_lat,row.centroid_lng)),
    axis=1
)
```
